```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import DensityMatrix, state_fidelity
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
import matplotlib.pyplot as plt
```

```python

# 1. Quantum state generation
def make_state(label):
    qc = QuantumCircuit(2)
    if label == 0:
        qc.h(0)
        qc.cx(0, 1)
    else:
        qc.x(0)
        qc.h(1)
    return qc


def get_statevector(qc, noise_model=None):
    """
    Gets the statevector of a quantum circuit.
    """
    # Create a copy of the circuit to avoid modifying the original
    qc_copy = qc.copy()
    qc_copy.save_statevector()
    if noise_model is None:
        backend = AerSimulator(method='statevector')
    else:
        backend = AerSimulator(method='statevector', noise_model=noise_model)

    tqc = transpile(qc_copy, backend)
    # Run the circuit; result will have a single experiment
    result = backend.run(tqc, shots=1).result()
    # Specify experiment index 0
    return result.get_statevector(0)

# Noise model
noise_model = NoiseModel()
# Define separate errors for single-qubit and two-qubit gates
single_qubit_error = depolarizing_error(0.05, 1)
two_qubit_error = depolarizing_error(0.05, 2)  # Error for 2-qubit gates

# Add errors to the noise model for their respective gates
noise_model.add_all_qubit_quantum_error(single_qubit_error, ['u1', 'u2', 'u3'])
noise_model.add_all_qubit_quantum_error(two_qubit_error, ['cx'])


# 3. Generate dataset
n_samples = 100
X_ideal, X_noisy, y, dm_ideal = [], [], [], []
for _ in range(n_samples):
    label = np.random.randint(0, 2)
    qc = make_state(label)
    sv_ideal = get_statevector(qc)
    sv_noisy = get_statevector(qc, noise_model)
    X_ideal.append(np.hstack([sv_ideal.real, sv_ideal.imag]))
    X_noisy.append(np.hstack([sv_noisy.real, sv_noisy.imag]))
    y.append(label)
    dm_ideal.append(DensityMatrix(sv_noisy).data)  # Use full complex density matrix

X_ideal = np.array(X_ideal, dtype=np.float32)
X_noisy = np.array(X_noisy, dtype=np.float32)
dm_ideal = np.array(dm_ideal, dtype=np.complex64)
y = np.array(y, dtype=np.int64)

# 4. Train/test split
X_ideal_train, X_ideal_test, X_noisy_train, X_noisy_test, dm_train, dm_test = train_test_split(
    X_ideal, X_noisy, dm_ideal, test_size=0.2, random_state=42)

# 5. CycleGAN components
class Generator(nn.Module):
    def __init__(self, dim=8):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim, 64), # Increased hidden layer size
            nn.ReLU(),
            nn.BatchNorm1d(64),  # Added batch normalization
            nn.Linear(64,64),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.Linear(64, dim)
        )
    def forward(self, x):
        return self.net(x)

class Discriminator(nn.Module):
    def __init__(self, dim=8):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim, 64),  # Increased hidden layer size
            nn.LeakyReLU(0.2),   # Changed ReLU to LeakyReLU
            nn.Dropout(0.25),    # Added dropout
            nn.Linear(64, 64),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.25),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.net(x)

# Store loss values for plotting
loss_g_history = []
loss_d_x_history = []
loss_d_y_history = []

device = 'cuda' if torch.cuda.is_available() else 'cpu'
X_ideal_train_torch = torch.tensor(X_ideal_train, device=device)
X_noisy_train_torch = torch.tensor(X_noisy_train, device=device)

G = Generator().to(device)  # Ideal -> Noisy
F = Generator().to(device)  # Noisy -> Ideal
D_X = Discriminator().to(device)
D_Y = Discriminator().to(device)

criterion_GAN = nn.BCELoss()
criterion_cycle = nn.L1Loss()
optimizer_G = optim.Adam(list(G.parameters()) + list(F.parameters()), lr=2e-4, betas=(0.5, 0.999)) # Adjusted betas
optimizer_D_X = optim.Adam(D_X.parameters(), lr=2e-4, betas=(0.5, 0.999))
optimizer_D_Y = optim.Adam(D_Y.parameters(), lr=2e-4, betas=(0.5, 0.999))

# 6. CycleGAN training
epochs = 100  # Increased number of epochs
batch_size = 64 # Increased batch size
lambda_cycle = 10 # Adjusted cycle loss weight

for epoch in range(epochs):
    perm = torch.randperm(X_ideal_train_torch.size(0))
    for i in range(0, X_ideal_train_torch.size(0), batch_size):
        idx = perm[i:i+batch_size]
        real_X = X_ideal_train_torch[idx]
        real_Y = X_noisy_train_torch[idx]
        valid = torch.ones((real_X.size(0), 1), device=device)
        fake = torch.zeros((real_X.size(0), 1), device=device)

        # Train Generators
        optimizer_G.zero_grad()
        fake_Y = G(real_X)
        loss_GAN_XY = criterion_GAN(D_Y(fake_Y), valid)
        fake_X = F(real_Y)
        loss_GAN_YX = criterion_GAN(D_X(fake_X), valid)
        recov_X = F(fake_Y)
        loss_cycle_X = criterion_cycle(recov_X, real_X)
        recov_Y = G(fake_X)
        loss_cycle_Y = criterion_cycle(recov_Y, real_Y)
        loss_G = loss_GAN_XY + loss_GAN_YX + lambda_cycle * (loss_cycle_X + loss_cycle_Y)
        loss_G.backward()
        optimizer_G.step()

        # Train Discriminators
        optimizer_D_Y.zero_grad()
        loss_real = criterion_GAN(D_Y(real_Y), valid)
        loss_fake = criterion_GAN(D_Y(fake_Y.detach()), fake)
        loss_D_Y = 0.5 * (loss_real + loss_fake)
        loss_D_Y.backward()
        optimizer_D_Y.step()

        optimizer_D_X.zero_grad()
        loss_real = criterion_GAN(D_X(real_X), valid)
        loss_fake = criterion_GAN(D_X(fake_X.detach()), fake)
        loss_D_X = 0.5 * (loss_real + loss_fake)
        loss_D_X.backward()
        optimizer_D_X.step()
    loss_g_history.append(loss_G.item())
    loss_d_x_history.append(loss_D_X.item())
    loss_d_y_history.append(loss_D_Y.item())
    print(f"Epoch {epoch+1}/{epochs}, Loss_G: {loss_G.item():.4f}, Loss_D_X: {loss_D_X.item():.4f}, Loss_D_Y: {loss_D_Y.item():.4f}")

# Plotting training losses
plt.figure(figsize=(10, 5))
plt.plot(loss_g_history, label='Generator Loss')
plt.plot(loss_d_x_history, label='Discriminator X Loss')
plt.plot(loss_d_y_history, label='Discriminator Y Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('CycleGAN Training Losses')
plt.legend()
plt.grid(True)
plt.savefig('cyclegan_training_losses.png')
plt.show()

# 7. Use G to create noisy state from ideal data (for train and test sets)

with torch.no_grad():
    noisy_simulated_train = G(X_ideal_train_torch).cpu().numpy()
    X_ideal_test_torch = torch.tensor(X_ideal_test, device=device)
    noisy_simulated_test = G(X_ideal_test_torch).cpu().numpy()
    denoised_test = F(X_noisy_train_torch).cpu().numpy()

# 8. Tomography model: reconstruct density matrix from statevector
# We'll output both real and imaginary parts (32 outputs)
class TomographyNN(nn.Module):
    def __init__(self, input_dim=8, output_dim=32):  # 8 for statevector, 32 for 4x4 complex density matrix
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128), nn.ReLU(),
            nn.Linear(128, 128), nn.ReLU(),
            nn.Linear(128, output_dim)
        )
    def forward(self, x):
        return self.net(x)

def prepare_dm_targets(dm_array):
    # Flatten real and imag parts for each density matrix
    return np.hstack([dm_array.real.reshape(len(dm_array), -1), dm_array.imag.reshape(len(dm_array), -1)]).astype(np.float32)

dm_train_flat = prepare_dm_targets(dm_train)
dm_test_flat = prepare_dm_targets(dm_test)
fidelity_history = []

def train_tomography_nn(X, y, epochs=10):
    X_torch = torch.tensor(X, dtype=torch.float32, device=device)
    y_torch = torch.tensor(y, dtype=torch.float32, device=device)
    dataset = TensorDataset(X_torch, y_torch)
    loader = DataLoader(dataset, batch_size=32, shuffle=True)
    model = TomographyNN(input_dim=X.shape[1], output_dim=y.shape[1]).to(device)
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    criterion = nn.MSELoss()

    loss_history = []  # Store loss values for plotting

    for epoch in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            out = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
        loss_history.append(loss.item())
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")
        current_fidelity = evaluate_tomography_nn(model, X, y)
#Evaluate fidelity
        fidelity_history.append(current_fidelity)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}, Fidelity: {current_fidelity:.4f}")
        if current_fidelity >= 0.50:
            print(f"Reached fidelity of 0.50 at epoch {epoch+1}!")
            break
    # Plotting training loss
    plt.figure(figsize=(8, 4))
    plt.plot(loss_history, label='TomographyNN Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('TomographyNN Training Loss')
    plt.legend()
    plt.grid(True)
    plt.savefig('tomography_nn_training_loss.png')
    plt.show()

    return model

def nearest_density_matrix(mat):
    # Make Hermitian
    mat = (mat + mat.T.conj()) / 2
    # Eigen-decomposition
    eigvals, eigvecs = np.linalg.eigh(mat)
    eigvals = np.maximum(eigvals, 0)
    mat = eigvecs @ np.diag(eigvals) @ eigvecs.conj().T
    mat /= np.trace(mat)
    return mat

def evaluate_tomography_nn(model, X_test, y_test):
    model.eval()
    X_torch = torch.tensor(X_test, dtype=torch.float32, device=device)
    with torch.no_grad():
        preds = model(X_torch).cpu().numpy()
    fidelities = []
    valid_count = 0  # Counter for valid density matrices
    for pred, true in zip(preds, y_test):
        try:
            # Reshape and ensure complex numbers
            pred_real = pred[:16].reshape((4,4))
            pred_imag = pred[16:].reshape((4,4))
            pred_dm = pred_real + 1j * pred_imag
            # Apply nearest density matrix function
            pred_dm = nearest_density_matrix(pred_dm)
            # Ensure true density matrix is correctly reshaped
            true_real = true[:16].reshape((4,4))
            true_imag = true[16:].reshape((4,4))
            true_dm = true_real + 1j * true_imag
            # Calculate fidelity
            fid = state_fidelity(DensityMatrix(pred_dm), DensityMatrix(true_dm))
            fidelities.append(fid)
            valid_count += 1  # Increment valid counter
        except Exception as e:
            print("Fidelity error:", e)
            # Skip adding fidelity if it's an invalid density matrix
            continue
    if valid_count > 0:
        avg_fid = np.mean(fidelities)
    else:
        avg_fid = 0.0  # If no valid density matrices, set fidelity to 0
    print(f"Average Fidelity: {avg_fid:.4f} (Valid count: {valid_count}/{len(X_test)})")
    return avg_fid

# 9. Train and evaluate tomography models
# --- Train on IDEAL data ---
print("\nTraining TomographyNN on IDEAL data:")
tomography_ideal = train_tomography_nn(X_ideal_train, dm_train_flat)

# --- Train on CycleGAN-augmented data ---
print("\nTraining TomographyNN on CycleGAN-augmented data:")
tomography_cgan = train_tomography_nn(noisy_simulated_train, dm_train_flat)

# --- Evaluate both on NOISY test set ---
print("\nEvaluating TomographyNN (IDEAL-trained) on NOISY test set:")
fidelity_ideal = evaluate_tomography_nn(tomography_ideal, X_noisy_test, dm_test_flat)

print("\nEvaluating TomographyNN (CycleGAN-augmented-trained) on NOISY test set:")
fidelity_cgan = evaluate_tomography_nn(tomography_cgan, X_noisy_test, dm_test_flat)

# --- Print Comparison ---
print("\n=== Fidelity Comparison on NOISY Test Set ===")
print(f"TomographyNN trained on IDEAL data:           {fidelity_ideal:.4f}")
print(f"TomographyNN trained on CycleGAN-augmented:   {fidelity_cgan:.4f}")

if fidelity_cgan > fidelity_ideal:
    print("CycleGAN-augmented training improves performance on noisy data!")
else:
    print("Ideal-only training performs better on noisy data (unexpected).")


# 10. Visualization of Density Matrices (example)
def visualize_density_matrix(dm, title="Density Matrix"):
    """Visualizes the real part of a density matrix."""
    plt.figure(figsize=(6, 5))
    plt.imshow(np.real(dm), cmap='viridis', interpolation='nearest')
    plt.title(title)
    plt.colorbar()
    plt.xticks(range(dm.shape[0]))
    plt.yticks(range(dm.shape[1]))
    plt.savefig(f'{title}.png')
    plt.show()

# Choose a random index from the test set to visualize
idx = np.random.randint(0, len(dm_test))

# Visualize the true density matrix
visualize_density_matrix(np.real(dm_test[idx]), title="True Density Matrix")

# Get predictions from the trained TomographyNN models
with torch.no_grad():
    ideal_pred = tomography_ideal(torch.tensor(X_noisy_test[idx:idx+1], dtype=torch.float32, device=device)).cpu().numpy()
     # sim_pred = tomography_sim(torch.tensor(X_noisy_test[idx:idx+1], dtype=torch.float32, device=device)).cpu().numpy()
    cgan_pred = tomography_cgan(torch.tensor(noisy_simulated_test[idx:idx+1], dtype=torch.float32, device=device)).cpu().numpy()

# Reshape the predicted density matrices
ideal_dm = nearest_density_matrix((ideal_pred[0][:16].reshape(4, 4) + 1j * ideal_pred[0][16:].reshape(4, 4)))
# sim_dm = nearest_density_matrix((sim_pred[0][:16].reshape(4, 4) + 1j * sim_pred[0][16:].reshape(4, 4)))
cgan_dm = nearest_density_matrix((cgan_pred[0][:16].reshape(4, 4) + 1j * cgan_pred[0][16:].reshape(4, 4)))

# Visualize the predicted density matrices

visualize_density_matrix(np.real(ideal_dm), title="Predicted Density Matrix (Ideal Trained)")
#visualize_density_matrix(np.real(sim_dm), title="Predicted Density Matrix (Noisy Trained)")
visualize_density_matrix(np.real(cgan_dm), title="Predicted Density Matrix (CycleGAN Trained)")
```
